```python
import pandas as pd

data = pd.read_csv( '/home/andrw/script/repository/ycbs255/jupyter/data/Credit.csv')
data.head(n=5)

credit = data[['Income','Limit','Rating','Cards','Age','Education','Balance']]

credit.head(n=2)
credit.describe()
```

```python
from pandas.plotting import scatter_matrix
%matplotlib inline
scatter_matrix(credit,alpha=0.2);
```

```python
#Correlation
import matplotlib.pyplot as plt
%matplotlib inline

corr = credit.corr()
fig,ax = plt.subplots(figsize=(10,10))
ax.matshow(corr)

plt.title('Correlation credit dataset')
plt.xticks(range(len(corr.columns)), corr.columns);
plt.yticks(range(len(corr.columns)), corr.columns);
```

```python
from sklearn.linear_model import Ridge
import numpy as np
y= credit['Balance'].values
X= credit[['Income','Limit','Rating','Cards','Age','Education']]
rr = Ridge(alpha = 100,normalize =True) # alpha ois the lambda that reduce the number of columns in the formula
rr.fit(X,y)

X_predict = np.array([15,3000,300,2,34,16]).reshape(1,6)
rr.predict(X_predict)
```
