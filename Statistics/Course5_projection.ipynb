```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

data = np.loadtxt('/home/andrw/script/repository/ycbs255/jupyter/data/ziptrain.csv')
print(data.shape)

plt.imshow(-data[2,1:].reshape(16,16),"gray");
```

```python
from sklearn.decomposition import PCA

zipdata3 = data[data[:,0] == 3]
zipdata8 = data[data[:,0] == 8]
zipdata38 = np.vstack([zipdata3 , zipdata8])

pca = PCA(n_components = 2)
pca.fit(zipdata38[:, 1:])
Z = pca.transform(zipdata38[:,1:])
plt.scatter(Z[: , 0],Z[: , 1], c = zipdata38[: , 0] , alpha = 0.3);
```
