```python
import pandas as pd
wage = pd.read_csv('/home/andrw/script/repository/McGill/data/Wage.csv')
wage.describe()

print(wage.head(n=2))
```

```python
from sklearn.preprocessing import PolynomialFeatures
import statsmodels.api as sm
pd.options.mode.chained_assignment = None #No warnings

X1 = PolynomialFeatures(1).fit_transform(wage[['age']])
X2 = PolynomialFeatures(2).fit_transform(wage[['age']])
X3 = PolynomialFeatures(3).fit_transform(wage[['age']])
X4 = PolynomialFeatures(4).fit_transform(wage[['age']])
X5 = PolynomialFeatures(5).fit_transform(wage[['age']])

for i in X4:
    print(i)

y=(wage['wage']>250).map({False:0,True:1})
lr = sm.GLM(y,X1,
           family=sm.families.Binomial(sm.families.links.logit))
lr=lr.fit()
lr.summary()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

age_grid = np.arange(20,300).reshape(-1,1)
X_test = PolynomialFeatures(1).fit_transform(age_grid)
y_hat = lr.predict(X_test)

plt.plot(age_grid, y_hat, color='b');
```
