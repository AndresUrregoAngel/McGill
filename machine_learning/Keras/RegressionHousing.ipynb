The goals is build up a model to predict the medias house value from a dataset tthat counts with few variables and the column indicated the avarage price based on each previous column. 


1. I will load the dataset from the repository and normalize the target column (6)

```

import pandas as pd
import random
from keras.datasets import mnist
import keras
from keras.layers import Dense
from keras.models import Sequential
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('https://raw.githubusercontent.com/nrkfeller/YCBS258/master/week4/data/cal_housing_clean.csv', skiprows=1,header=None)
df.head(n=5)
```

1.1 Based in the load dataset the intention is predict the column **medianHouseValue**, this column must be normalized between  `0` and `1` 

```
df[6]=((df[6]-df[6].min())/(df[6].max()-df[6].min()))*1
print(df[6].head(n=5)) #sample of normalized column

print(df.head(n=5))
```

2. Using  `Sklearn` I spread out my data in the testing and training samples to build up the model.

```
from sklearn.model_selection import train_test_split, cross_val_score

y = df[6]  # Define the target
X = df.drop([6],axis=1) # Define the variables for the model

print(X.head(n=5))

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.3, shuffle=False)


print(X_train[0].shape)
```

3. I move forwaard with the model creation, therefore I define a first layer to receive the first input and set the variable activation with **sgmoid**. Besides I include 2 extra layer to enhance my model and the last layer has only a single neuron due to the expected output is just one value.

```
# create model
model = Sequential()
model.add(Dense(20, input_shape=(6,), kernel_initializer='normal', activation='sigmoid'))
model.add(Dense(32,activation='relu'))
model.add(Dense(32,activation='relu'))
model.add(Dense(1,  activation='sigmoid'))


model.compile(loss='mean_squared_error',
              optimizer='adam',)
             # metrics=['accuracy'])
```

4 . Finally I fit my model with the train dataset, I assign the batch size how this data will break down in **10** and applying to each chunk 20 epochs.
 

```

h = model.fit(X_train, y_train, batch_size=10, epochs=20,
               verbose=1, validation_split=0.3)

```

5. I can render a summary of my model summing up the model layers.

```
model.summary()
```
